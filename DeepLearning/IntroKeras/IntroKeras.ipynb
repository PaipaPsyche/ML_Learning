```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,confusion_matrix
```

### Importando los datos
Estos datos los genere yo a partir de unas reglas matematicas establecidas sobre caracteristicas de las palabras. La caracteristica a predecir es si es un nombre propio o no.

```python
DATA=pd.read_csv("listado_clasificacion.csv")

H=list(DATA.keys().values)
H=list(filter(lambda x : x != 'propio' and x!="palabra", H))

features=DATA[H]
target=DATA[["propio"]]

X=features.values
Y=target.values

dim_=len(X[0])

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.5)
x_val = x_train[300:,]
y_val = y_train[300:,]
```

### Estructura de la Neural Network

```python
##Definiendo la arquitectura de la red.

model=Sequential()
model.add(Dense(16 ,input_dim=dim_,activation = "relu"))
model.add(Dense(8,activation = "relu"))
#model.add(Dropout(rate = 0.05,seed=100))
model.add(Dense(4,activation = "relu"))
model.add(Dense(1,activation = "hard_sigmoid")) 
```

```python
#configurando el modelo
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Entrenando...
model.fit(x_train, y_train, batch_size=50, epochs=200,validation_data=(x_val,y_val))
```

```python
results = model.evaluate(x_test, y_test)
for i in range(len(model.metrics_names)):
    print(model.metrics_names[i]," : ", results[i])
```

```python
model.predict(np.array([9,0,0,0,4,1,3,0,5,2,-4,0,1,0,3,0,0,1,0,0,1,0,0]).reshape(1,-1))
```

```python
y_pred=np.round(model.predict(x_test),0)
```

```python
np.sum(np.ones(len(y_pred)).reshape(-1,1)[y_test==1])
```

```python
confusion_matrix(y_test,y_pred)
```
